```python
shop_food = shop.loc[shop['상권업종대분류명']=='음식']
shop_food_DJ = shop_food.loc[shop_food['법정동명']=='두정동']


(ggplot(shop_food_DJ)
+ aes(x='경도', y='위도', color='상권업종중분류명')
+ geom_point(alpha=0.5, size=1.2)
+ theme(text=element_text(family='gulim'))
+ scale_fill_gradient(low = 'blue', high = 'green')
)
```

```python
import folium

map_osm = folium.Map(location=[36.8324651,127.137189], zoom_start=16)

for r in shop_food_DJ :
    if r[18]=='두정동' :
        if '음식' in r[4] :
            folium.Marker([float(r[-2]), float(r[-1])]).add_to(map_osm)
            map_osm.save('d:/dev/map3.html')
```
